```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.violinplot(x=df['New cases'], color='cyan')
plt.title("Violin Plot of New Cases")
plt.xlabel("New cases")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
states = df['Name of State / UT']
total_confirmed_cases = df['Total Confirmed cases']

# Create a horizontal bar chart with states on the y-axis
plt.figure(figsize=(10, 6))  # Optional: Set the figure size
plt.barh(states, total_confirmed_cases, color='cyan')
plt.ylabel('States / UT')  # Y-axis label
plt.xlabel('Total Confirmed Cases')  # X-axis label
plt.title('COVID-19 Total Confirmed Cases by State / UT')

# Display the graph
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter the data for June 13, 2020
filtered_data = df[df["Date"] == "13-06-2020"]

# Create a histogram without specifying custom bin edges
plt.figure(figsize=(10, 6))
plt.hist(filtered_data["Total Confirmed cases"], color='lightblue', edgecolor='black')
plt.title('Total Confirmed Cases Histogram (13-06-2020)')
plt.xlabel('Total Confirmed Cases')
plt.ylabel('Frequency')
plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))  # Adjust the figure size if needed
sns.barplot(x='Name of State / UT', y='Total Confirmed cases', data=df)
plt.grid(False)
plt.xticks(rotation=75)  # Rotate x-axis labels by 45 degrees for better readability
plt.xlabel('State / UT')
plt.ylabel('Total Confirmed Cases')
plt.title('Top 15 States with Highest Total Confirmed Cases')
plt.show()

```

```python

import pandas as pd
import matplotlib.pyplot as plt

# Group the data by 'Date' and sum 'New cases' and 'Death'
grouped_data = df.groupby('Date')[['New cases', 'Death']].sum()

# Create a line plot
plt.figure(figsize=(12, 6))
plt.plot(grouped_data.index, grouped_data['New cases'], label='New Cases', marker='o')
plt.plot(grouped_data.index, grouped_data['Death'], label='Deaths', marker='o')

# Customize the plot
plt.xlabel('Date')
plt.ylabel('Count')
plt.title('COVID-19 New Cases and Deaths Over Time')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)

# Show the plot
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Total Confirmed cases', y='Cured/Discharged/Migrated', data=df, hue='Name of State / UT')

# Customize the plot
plt.title('Scatter Plot: Total Confirmed cases vs. Cured/Discharged/Migrated')
plt.xlabel('Total Confirmed cases')
plt.ylabel('Cured/Discharged/Migrated')

# Show the plot
plt.grid()
plt.tight_layout()
plt.legend(title='State/UT', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a bivariate stacked bar chart
plt.figure(figsize=(12, 6))
plt.bar(df['Name of State / UT'], df['New cases'], label='New cases', color='skyblue',bottom=df['New recovered'])
plt.bar(df['Name of State / UT'], df['New recovered'], label='New recovered', color='lightgreen')

# Customize the plot
plt.title('Stacked Bar Chart: New Cases vs. New Recovered by State/UT')
plt.xlabel('State/UT')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Select the state you want to plot
selected_state = 'Maharashtra'
state_data = df[df['Name of State / UT'] == selected_state]

plt.figure(figsize=(12, 6))
plt.stackplot(state_data['Date'], state_data['Cured/Discharged/Migrated'], state_data['Total Confirmed cases'],
              labels=['Cured/Discharged/Migrated', 'Total Confirmed cases'], colors=['blue', 'lightgreen'], alpha=0.7)

plt.title(f'Stacked Area Chart: Daily Trends in {selected_state}')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend(loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
Category = ['Cured/Discharged/Migrated', 'deaths', 'Total Confirmed cases']
Counts = [df['Cured/Discharged/Migrated'].sum(), df['Death'].sum(), df['Total Confirmed cases'].sum()]
colors = ['pink','lightblue','lightgreen']
explode = (0.1, 0, 0)
plt.figure(figsize=(6, 6))
plt.pie(Counts, labels=Category, autopct='%1.1f%%', explode=explode, colors=colors, shadow=True, startangle=90)
plt.title('COVID-19 Distribution')
plt.axis('equal')
plt.show()


```
